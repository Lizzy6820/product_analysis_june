```python
import pandas as pd
import glob
import os
from dateutil.parser import parse as pr
import seaborn as sns
import matplotlib.pyplot as plt

path = r"C:\Users\DELL E6400\Downloads\product_data"
all_files = glob.glob(os.path.join(path, "*.csv"))     

df_from_each_file = (pd.read_csv(f) for f in all_files)
df  = pd.concat(df_from_each_file, ignore_index=True)
df
```

```python
#check for missing values
df.isnull().sum()
```

```python
#drop rows with misssing values
df = df.dropna()
df.isnull().sum()
```

```python
#create column for amount per product
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors = 'coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'], errors = 'coerce')
df['Amount_Sold'] = df['Price Each'] * df['Quantity Ordered']

#Breakdown the addresses into city, state and postalcode
#df['City'] = df['Purchase Address'].apply(lambda x: x.split(' ')[0].split(',')[1])
df['State'] = df['Purchase Address'].apply(lambda x: x.split()[-2])
df['Postal Code'] = df['Purchase Address'].apply(lambda x: x.split()[-1])
df.head()
```

```python
#Breakdown order time into year, month and hour
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
df['Hour'] = df['Order Date'].dt.hour
df.head()

```

```python
#total annual sales per month
months={1.0:'Jan', 2.0:'Feb', 3.0:'Mar', 4.0:'Apr',5.0:'May',6.0:'Jun',7.0:'Jul',8.0:'Aug', 9.0:'Sept',10.0:'0ct',11.0:'Nov', 12.0:'Dec'}
df2 = df.replace({'Month': months})
#print (df2.head())
total_annual_sales = df2.groupby(['Month']).Amount_Sold.sum().sort_values(ascending=True)
print(total_annual_sales)
total_annual_sales.plot(kind='bar', color=(0.2, 0.4, 0.6, 0.6),figsize=(16, 8))
plt.title('Total annual sales per month')
plt.ylabel('Total annual sales')

plt.show()

```

```python
#correlation matrix
plt.figure(figsize=(24, 8))
sns.heatmap(df.corr(), annot=True);
```

```python
plt.figure(figsize=(24, 6))
sns.barplot(
    x=df['Month'], 
    y=df['Amount_Sold'], 
    data=total_annual_sales, palette='deep'
)
# title
plt.title('Month Sales and Much Earned in that Months', 
          fontname='monospace', weight='bold'
)
# x-label
plt.xlabel('Months')
# y-label
plt.ylabel('Sales in USD ($)');
```

```python
#States with highest number of sales
states= df.groupby(['State']).Amount_Sold.sum()
states.plot(kind='bar', color=(0.2, 0.4, 0.6, 0.6),figsize=(16, 8))
plt.title('States with highest number of sales')
plt.ylabel('Total annual sales')
plt.show()

```

```python
#Hours with the largest quantity of orders
orders= df.groupby(['Hour'])['Quantity Ordered'].sum()
orders.plot(color=(0.2, 0.4, 0.6, 0.6),figsize=(16, 4))
plt.style.use('seaborn-whitegrid')
plt.title('Hours with highest number of orders')
plt.ylabel('Total number of orders')
plt.show()
```
